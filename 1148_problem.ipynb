```python
data = {"article_id":[1,1,2,2,4,3,3],"author_id":[3,3,7,7,7,4,4],"viewer_id":[5,6,7,6,1,4,4],"view_date":["2019-08-01","2019-08-02","2019-08-01","2019-08-02","2019-07-22","2019-07-21","2019-07-21"]}
```

```python
views = pd.DataFrame(data)
```

```python
#Write a solution to find all the authors that viewed at least one of their own articles.
```

```python
    # filter rows where author_id == viewer_id
filter_dataset = views[views['author_id'] == views['viewer_id']]
```

```python
 # Get unique author_ids from the filtered data
df = filter_dataset['author_id'].unique()
```

```python

   # Sort the unique author_ids in ascending order
   df = sorted(df)
```
